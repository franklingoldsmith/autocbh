### Create YAML Files

```python
nrg = pd.read_pickle('./data/energies_Franklin.pkl')
nrg
```

```python
from itertools import compress
# keys we disregard
constants = ['R', 'kB', 'h', 'c', 'amu', 'GHz_to_Hz', 'invcm_to_invm', 'P_ref', 
             'hartree_to_kcalpermole', 'hartree_to_kJpermole', 'kcalpermole_to_kJpermole', 'uncertainty']
# necessary keys for theories 
theory = {'anl0' : ['avqz', 'av5z', 'zpe', 'ci_DK', 'ci_NREL', 'core_0_tz', 'core_X_tz', 'core_0_qz',
                    'core_X_qz', 'ccT', 'ccQ', 'zpe_harm', 'zpe_anharm'],
          'b2plypd3': ['b2plypd3_E0'],
          'f12b': ['b2plypd3_zpe', 'f12b'],
          'm062x_dnlpo': ['m062x_zpe', 'm062x_dnlpo'],
          'wb97xd_dnlpo': ['wb97xd_zpe', 'wb97xd_dnlpo'],
          'm062x' : ['m062x_E0'], 
          'wb97xd' : ['wb97xd_E0']}
# load dataframe of energies
nrg = pd.read_pickle('./data/energies_Franklin.pkl').drop(constants,axis=1)
nrg[['avqz', 'av5z', 'zpe', 'ci_DK', 'ci_NREL', 'core_0_tz',
       'core_X_tz', 'core_0_qz', 'core_X_qz', 'ccT', 'ccQ', 'zpe_harm',
       'zpe_anharm', 'b2plypd3_zpe', 'b2plypd3_E0', 'f12a', 'f12b',
       'm062x_zpe', 'm062x_E0', 'm062x_dnlpo', 'wb97xd_zpe', 'wb97xd_E0',
       'wb97xd_dnlpo']] = nrg[['avqz', 'av5z', 'zpe', 'ci_DK', 'ci_NREL', 'core_0_tz',
       'core_X_tz', 'core_0_qz', 'core_X_qz', 'ccT', 'ccQ', 'zpe_harm',
       'zpe_anharm', 'b2plypd3_zpe', 'b2plypd3_E0', 'f12a', 'f12b',
       'm062x_zpe', 'm062x_E0', 'm062x_dnlpo', 'wb97xd_zpe', 'wb97xd_E0',
       'wb97xd_dnlpo']].replace(0, np.nan)

# Generate and format YAML files for each molecule
na_sources = nrg.loc[:,'source'].isna()
nrg.loc[na_sources,'DfH'] = np.nan
for m in nrg.index:
    d = {}
    d['smiles'] = CanonSmiles(m)
    d['alias'] = [nrg.loc[m,'alias']]
    d['theory'] = {}
    if not na_sources[m]:
        d['heat_of_formation'] = {nrg.loc[m,'source'] : float(nrg.loc[m, 'DfH'])}
    for method in theory:
#         # heat of formation for ATcT
#         if not na_sources[m]:
#             d['theory']['ATcT'] = {'heat_of_formation' : float(nrg.loc[m, 'DfH'])}
            
        d['theory'][method] = nrg.loc[m, list(compress(theory[method], 
                                                       ~nrg.loc[m, theory[method]].isna().values))].to_dict()
    
    filename = f'./data/molecule_data/{CanonSmiles(m)}.yaml'
    with open(filename, 'w') as f:
        yaml.dump(d, f, default_flow_style=False)
        
out = molData.generate_database('data/molecule_data/')

# Generate methods_keys.yaml file
with open('./data/methods_keys.yaml','w') as f:
    yaml.dump(out[1], f, default_flow_style=False)
```
