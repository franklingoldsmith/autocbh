```python
df = pd.read_excel('../../PFAS Energies/PFAS Energies f2.xlsx')
```

```python
start_ind = df[df['i'] == 'RAW ENERGIES'].index
end_ind = df[df['i'] == 'Heats of Formation (0K)'].index
smiles_ind = np.where(df.iloc[start_ind+2].values=='smiles')[1][0]

nrg = df[start_ind[0]+2:end_ind[0]-2].drop(df.iloc[:,smiles_ind+1:],axis=1)
nrg.columns = nrg.iloc[0,:]
nrg = nrg.iloc[1:,:]
nrg.name = 'Raw Energies'
nrg.columns.name = None
nrg.set_index('smiles', inplace=True)
# nrg.fillna(0, inplace=True)
ref_atct = df.iloc[5:12,[0,2]].set_index('i')
```

```python
idx_names = [
    ['dft', 'dft', 'dft', 'dft', 'dft', 'dft', 'dft', 'dft', 'dft',
              'f12', 'f12', 'f12', 'f12', 'f12', 'f12', 'f12', 'f12', 'f12',
              'dnlpo', 'dnlpo', 'dnlpo', 'dnlpo', 'dnlpo', 'dnlpo', 'dnlpo', 'dnlpo', 'dnlpo'], 
    ['wb97xd', 'wb97xd','wb97xd', 'm062x', 'm062x', 'm062x', 'b3lyp', 'b3lyp', 'b3lyp', 
     'wb97xd', 'wb97xd','wb97xd', 'm062x', 'm062x', 'm062x', 'b3lyp', 'b3lyp', 'b3lyp',
     'wb97xd', 'wb97xd','wb97xd', 'm062x', 'm062x', 'm062x', 'b3lyp', 'b3lyp', 'b3lyp'],
    ['e', 'zpe', 'pdt_freq', 'e', 'zpe', 'pdt_freq', 'e', 'zpe', 'pdt_freq', 
    'e0', 'zpe', 't1', 'e0', 'zpe', 't1', 'e0', 'zpe', 't1', 
    'e0', 'zpe', 't1', 'e0', 'zpe', 't1', 'e0', 'zpe', 't1']]
indxs = pd.Index(list(zip(*idx_names)))
```

```python
energies = nrg.T.drop(['Formula', 'Species']).set_index(indxs).T.drop(np.NaN)
# standardize smiles
smiles = [Chem.MolToSmiles(Chem.MolFromSmiles(smile)) for smile in energies.index]
energies.index = smiles
```
